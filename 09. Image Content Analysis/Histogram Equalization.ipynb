```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load image
img = cv2.imread('sunrise.jpg')
```

```python
%matplotlib inline

# Convert it to grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img_gray, cmap='gray')
plt.show()
```

```python
# Equalize the histogram
img_gray_histeq = cv2.equalizeHist(img_gray)

plt.imshow(img_gray_histeq, cmap='gray')
plt.show()
```

```python
# Histrogram equalization of color images
img_yuv = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)
```

```python
img_yuv[:,:,0] = cv2.equalizeHist(img_yuv[:,:,0])
```

```python
img_histeq = cv2.cvtColor(img_yuv, cv2.COLOR_YUV2BGR)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
plt.imshow(cv2.cvtColor(img_histeq, cv2.COLOR_BGR2RGB))
plt.show()
```
