```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load an image
img = cv2.imread('forest.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# Show image
%matplotlib inline
# cv2.imshow('Original', img)

plt.imshow(img)
plt.show()
```

```python
# Cropping Image
h, w, = img.shape[:2]
start_row, end_row = int(0.21*h), int(0.73*h)
start_col, end_col = int(0.37*w), int(0.92*w)

img_cropped = img[start_row:end_row, start_col:end_col]
plt.imshow(img_cropped)
plt.show()
```

```python
# Resizing an image
scaling_factor = 1.3
img_scaled = cv2.resize(img, None, fx=scaling_factor, 
                        fy=scaling_factor, 
                        interpolation=cv2.INTER_LINEAR)
plt.imshow(img_scaled)
plt.show()
```

```python
# Skew an image
img_skewed = cv2.resize(img, (250,400),
                       interpolation=cv2.INTER_AREA)
plt.imshow(img_skewed)
plt.show()
```

```python
# Save an image
output_file = 'forest_cropped.jpg'
img_save = cv2.cvtColor(img_cropped, cv2.COLOR_BGR2RGB)
cv2.imwrite(output_file, img_save)
```
